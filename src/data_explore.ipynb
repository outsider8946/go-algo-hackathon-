#**Частотная статистика по тикерам внутри исходных файлов с данными**

```python
path_to_data = '/content/drive/MyDrive/Colab Notebooks/goalgo_hac/data/tradestats/'
all_tradestats = ['tradestats_2020.csv','tradestats_2021.csv','tradestats_2022.csv','tradestats_2023.csv']
frames = []
index = 0
for stat in all_tradestats:
  print(stat)
  df = pd.read_csv(path_to_data + stat,sep = ';')
  frames.append(df)
  df.index = range(index,index + len(df))
  index = len(df)+1
all_data = pd.concat(frames)
```

```python
all_data['secid'].value_counts().sort_values(ascending=False).iloc[0:20]
```
